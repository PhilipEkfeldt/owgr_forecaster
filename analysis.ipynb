```python
import pandas as pd;
import numpy as np;
import datetime;
```

```python
data = pd.read_csv('scraped.csv')
data.head()
```

```python
def sunday_date(week_year_str):
    return datetime.datetime.strptime(week_year_str + '-0', "%Y-W%W-%w")

data['date_sun'] = ( data['year'].astype(str) + "-W" + data['week'].astype(str) ).map(sunday_date)

data['weeks_since_today'] = ( datetime.datetime.today() - data['date_sun']  ).dt.days // 7

data['coefficient'] = 1 - np.maximum( data['weeks_since_today'] - 12, 0 ) /92
```
